```python
from sklearn.model_selection import train_test_split
```

```python
y = df.Attrition
df.drop(['Attrition'], axis=1, inplace=True)
df
```

```python

dummies = pd.get_dummies(df, columns=['BusinessTravel','Department','EducationField','JobRole', 'Gender', 'MaritalStatus'],prefix='OH')
dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_normalized = scaler.fit_transform(dummies)
x_normalized
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_normalized, y, test_size=0.25, random_state=0, shuffle = True)
```

```python
x_train
```

```python

x_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000, C=100)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_prob = model.predict_proba(x_test)
y_prob.shape
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python

print(classification_report(y_test,y_pred))
```

```python

model = LogisticRegression(max_iter=200, solver='newton-cg', class_weight='balanced')
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print('accuracy:', round(accuracy_score(y_test,y_pred)*100,3),'%')
```

```python
def plot_top_features(importance, features, model_type): 
    feature_importance = np.array(importance)
    feature_names = np.array(features)
    
    data={'feature_names':feature_names,'feature_importance':abs(feature_importance),'sign': np.sign(feature_importance)}
    table = pd.DataFrame(data) 
    table.sort_values(by=['feature_importance'], ascending=False,inplace=True) 
    
    plt.figure(figsize=(10,20))
    sns.barplot(x=table['feature_importance'][:15], y=table['feature_names'][:15], hue=table['sign'][:15])
    plt.title(model_type + ' FEATURE IMPORTANCE')
    plt.xlabel('Feature importance')
    plt.ylabel('Features')

plot_top_features(model.coef_[0,:], dummies.columns,'Logistic Regression')
```
